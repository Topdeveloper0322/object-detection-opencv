# 1. Using Images with cv2

```python
import numpy
import matplotlib.pyplot as plt
import matplotlib.image as img 
import cv2
```

## Import local images

Import local image using cv2 method `imread`.
* Import image and convert it to grayscale.
* Import directly grayscale image. 

**_good practice_**: use an absolute path to the image.
*Note*: Remember that `imgread` take the image using BGR as opposed to RGB. [see first plot]

```python
#Get the img matrix
image = cv2.imread("imgs/paradrop.png")
#Convert the image to gray
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#Import directly grayscale image
direct_gray_image = cv2.imread("imgs/paradrop.png", cv2.IMREAD_GRAYSCALE)
```

Exercise: 

Represent images using `matplotlib`. It is possible to obtain pixels values and coordinates.

```python
#Represent images using matplotlib
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
axs[0].imshow(image)
axs[0].set_title("Original img")
axs[1].imshow(gray_image, cmap="gray")
axs[1].set_title("Convertet to gray")
axs[2].imshow(direct_gray_image, cmap="gray")
axs[2].set_title("Directly convertet to gray")
plt.show()
```

As commented below, `color_image` is represented taking RGB instead of BGR. Time to change it:

Note the slightly different colors perceived in `image_gray` and `image_converted_gray`. The good one is the converted one, as it was converted in RGB.

Exercise:


Plot the new images, coloured RGB, BGR converted to gray and rgb converted to gray, and see the diferences.

```python
#Change BGR to RGB 
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#Set subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
#Original image on RGB
axs[0].imshow(image_RGB)
axs[0].set_title("Original img")
#Bad image on BGR
axs[1].imshow(gray_image, cmap="gray")
axs[1].set_title("Converted to gray")
#Good image converted to gray from RGB
converted_to_gray_RGB = cv2.cvtColor(image_RGB, cv2.COLOR_BGR2GRAY)
axs[2].imshow(converted_to_gray_RGB, cmap="gray")
axs[2].set_title("Directly converted to gray")
plt.show()
```

## Write the images

In order to write the images is important to note that `cv2` method `imwrite` changes the input image to RGB as `cv2` thinks most of time it is better to write this way.

So in order to write the images correctly It is necessary to have the input image as BGR.

Exercise:
Save the three images.

```python
# Convert to BGR
image_BGR = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2BGR)
# Save images
cv2.imwrite("original_img.png", image_BGR)
cv2.imwrite("gray_img.png", gray_image)
cv2.imwrite("converted_to_gray_RGB.png", converted_to_gray_RGB)
```

## Wrap up

Following points need to be clear in your mind to be able to advance in `cv2`:

* Importing images:
    *  Convert from BGR to RGB.
    
* Process images: 
    *  Do it with RGB image.
   
* Write images:
    *  Convert RGB to BGR in order to write it (`cv2` automatically converts BGR to RGB).
